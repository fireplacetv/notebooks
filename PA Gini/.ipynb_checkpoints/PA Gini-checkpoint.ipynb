```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
players = pd.read_csv('data/2017.csv')
players.drop(0,inplace=True)

players['Salary'] = players['Salary'].replace('[\$,]','',regex=True).fillna(535000).astype(int)
players['Name'].replace('[*#\\\\].*','',regex=True, inplace=True)
players['PA'].fillna(0)
players['Salary per PA'] = players['Salary'] / players['PA']
players.reset_index(drop=True,inplace=True)
print(players.dtypes)

players.head()
```

```python
pd.pivot_table(players,
               'Tm',
               'Name',
               aggfunc='count').sort_values('Tm', ascending=False).head()
```

```python
sfg = players.loc[players['Tm']=='SFG'].replace(np.inf,0).sort_values('Salary per PA').copy()
```
